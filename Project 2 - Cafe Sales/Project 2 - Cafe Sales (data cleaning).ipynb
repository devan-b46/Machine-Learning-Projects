```python
df = pd.read_csv('/kaggle/input/cafe-sales-dirty-data-for-cleaning-training/dirty_cafe_sales.csv')
df
```

```python
# Let us change column names, to consistent way:
df.columns = ['transaction_id','item','quantity','price_per_unit',
              'total_spent','payment_method','location','transaction_date']
df
```

```python
# Checking for duplicates in the dataset

df.duplicated().sum()
```

```python
# Changing the Data Types of columns as needed:


df.quantity = df.quantity.astype('int')
df.price_per_unit = df.price_per_unit.astype('int')
df.total_spent = df.total_spent.astype('int')
df.transaction_date = pd.to_datetime(df.transaction_date,format='%Y-%m-%d')

df
```
